##### - View and delete function was edited to display error message if user typed in characters instead of integer
##### - Add and delete function was edited to write changes in data into the csv file

```python
# Author: Sejay Noupin
# Purpose: The purpose of this program is to enhance the
# Purpose: Game Manager program to store the data as a csv file

import csv

FILENAME = "game_manager.csv"

def write_games(game_list) :
    """
    The purpose of this function is to write incoming data
    from the game manager into the csv file
    """
    with open(FILENAME, "w", newline = "") as file :
        writer = csv.writer(file)
        writer.writerows(game_list)

def read_games() :
    """
    The purpose of this function is to read in the game manager file
    """
    try :
        game_list = []
        with open(FILENAME, newline = "") as file :
            reader = csv.reader(file)
            for row in reader :
                game_list.append(row)
        return game_list
    except FileNotFoundError :
        print("\nCould not find " + FILENAME + " file.")
        game_list = [["Mario Kart 8", "gif@email.com", "654,321"], 
                 ["Super Smash Brothers", "meme@me.com", "23,458"], 
                 ["Fortnite", "beast@ninja.com", "74,351"]]
        write_games(game_list)
        game_list = read_games()
        print("A starting list has been created, you can now continue with the manager")
        
def display_menu() :
    """
    This function runs the program to display the
    menu and the choices you are able to make
    """
    print("COMMAND MENU")
    print("list - Display all games")
    print("view - View a game")
    print("add - Add a game")
    print("del - Delete a game")
    print("exit - Exit program")

def list(game_list) :
    """
    This function runs the program to display
    the games that are already in the list
    """
    game_list = [] 
    game_list = read_games()
    if len(game_list) == 0 :
        print("There are no games in this list")
    else :
        i = 1
        for row in game_list :
            print(str(i) + ". " + row[0])
            i += 1
    print()
    
def view(game_list) :
    """
    This function shows the email of the high scorer
    and the high score of the selected game
    """
    game_list = [] 
    game_list = read_games()
    bad_data = True
    while bad_data :
        try :
            number = int(input("Number:"))
            while number < 1 or number > len(game_list) :
                print("Invalid game number")
                number = int(input("Number:"))
            bad_data = False
        except ValueError :
            print("Please enter a valid number. Letters are not acceptable.")
            
    print("Game: " + game_list[number - 1][0])
    print("High Scorer Email: " + game_list[number - 1][1])
    print("High Score: " + game_list[number - 1][2])
    print()
    
def add(game_list) :
    """
    This function runs the program to 
    add another game to the list
    """
    name = input("Game:")
    high_scorer = input("High Scorer Email:")
    high_score = (input("High Score:"))
    game = []
    game.append(name)
    game.append(high_scorer)
    game.append(high_score)
    game_list.append(game)
    print(game[0] + " was added.\n")

    write_games(game_list)

def delete(game_list) :
    """
    This function runs the program 
    to delete a movie from the list
    """
    bad_data = True
    while bad_data :
        try :
            number = int(input("Number:"))
            while number < 1 or number > len(game_list) :
                print("Invalid game number")
                number = int(input("Number:"))
            bad_data = False
        except ValueError :
                print("Please enter a valid number. Letters are not acceptable.")           
    
    game = game_list.pop(number-1)
    print(game[0] + " was deleted.")
    print()
    
    write_games(game_list)

def main() :
    """
    This is the main function that is called 
    in order to run the entire program
    """ 
    display_menu()
    game_list = []
    game_list = read_games()
    command = input("\nCommand:")

    while command.lower() != "exit" :
        if command.lower() == "list" :
            list(game_list)
        elif command.lower() == "view" :
            view(game_list)
        elif command.lower() == "add" :
            add(game_list)
        elif command.lower() == "del" :
            delete(game_list)
        else :
            print("Not a valid command. Please try again\n")
        display_menu()
        command = input("\nCommand:")
    print("Bye!")

if __name__ == "__main__" :
    main()
```
